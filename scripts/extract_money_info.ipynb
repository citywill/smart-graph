```python
import os

file_path = './docs/民间借贷/匡晓东与长春隆都房地产开发有限公司借款合同纠纷一审民事判决书.md'

content = open(file_path, 'r', encoding = 'utf - 8').read()
```

```python
prompt = f"""请提取文书属性：
- court：审理法院
- case_type：案件类型（如刑事、民事、行政等）
- time：裁判时间
- cause：案由（民事案由或刑事罪名等）
- defendants：被告（列表）
- plaintiffs：原告（列表）

使用中文回复。

严格按照JSON格式输出。
"""

import ask_llm

result = ask_llm.ask_llm(prompt,content[:1000])

print(f"回复：{result}")
```

```python
import split2chunks

chunks = split2chunks.split2chunks(content)

chunks
```

```python
prompt = f"""请在该裁判文书中，提出取借贷关系。
请用json返回，将关系列表存入"relations"字段。
关系是字典结构：
- subject：借贷主体，出借人
- object：借贷客体，借贷人
- amount：借贷金额
"""

for chunk in chunks:
    result = ask_llm.ask_llm(prompt,chunk)
    print(f"\n===== 文本：{chunk}")
    print(f"\n===== 结果：{result}")
```
